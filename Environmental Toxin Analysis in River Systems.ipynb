```python
# Check for missiing values
df_is_null = df.isnull().sum()
print(df_is_null)
```

```python
# fill missing values
numeric_columns = df.select_dtypes(include=[np.number]).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
```

```python
# Check for missiing values
df_is_null2 = df.isnull().sum()
print(df_is_null2)
```

```python
# Convert the 'Date' column to datatime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
average_toxin_levels = df[['Lead', 'Mercury', 'Arsenic']].mean()
print(f"Average Toxin Levels: \n{average_toxin_levels}")
```

```python
type(average_toxin_levels)
```

```python
# Identify top polluted rivers by avwerage lead levels
top_polluted_rivers = df.groupby('River_System')['Lead'].mean().sort_values(ascending=False)
print(f"Top Polluted Rivers by Lead Levels:\n {top_polluted_rivers}")
```

```python
type(top_polluted_rivers)
```

```python
# Plor Bar chart for average Lead Levels per river
plt.figure(figsize=(10, 6))
top_polluted_rivers.plot(kind='bar', color='skyblue')
plt.title('Average Lead Levels by River System')
plt.xlabel('River System')
plt.ylabel('Average Lead Levels')
plt.show()
```

```python
# Plot line graph for toxins levels over time
plt.figure(figsize=(10, 6))

sns.lineplot(x='Date', y='Lead', data=df, hue='River_System', errorbar=None)
plt.title('Lead Levels over time')
plt.xlabel('Date')
plt.ylabel('Lead Levels')
plt.show()
```

```python
df.columns
```

```python
# Compute the correlation matrix using only numeric columns
numeric_columns = ['pH_Level', 'Temperature', 'Lead', 'Mercury',
       'Arsenic', 'Nitrates', 'Phosphates', 'Dissolved_Oxygen']
corr_matrix = df[numeric_columns].corr()
```

```python
# Heatmap of correlation between different paraeters
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
# T-test between two river systems (e.g. Amazon and Nile)
amazon_data = df[df['River_System'] == 'Amazon']['Lead']
nile_data = df[df['River_System'] == 'Nile']['Lead']
t_stat, p_value = stats.ttest_ind(amazon_data, nile_data)
```

```python
print(f"T-test between Amazon and Nile: T-Statistic = {t_stat}, P-value = {p_value}")
```

```python
# Linear Regression analysis for toxin levels and pH
X = df[['pH_Level']]
y = df['Lead']
model = LinearRegression()
model.fit(X, y)
print(f"Linear Regression Coefficients: {model.coef_}")
print(f"Linear Regression Intercept: {model.intercept_}")
```
